```python
import numpy as np
import math

def sign(x):
    x = np.sign(x)
    if len(x.shape) == 0:
        if x == 0:
            x = -1
        return x
    for i in range(x.shape[0]):
        if x[i] == 0:
            x[i] = -1
    return x

n = 200
x = 3 * (np.random.rand(n, 4) - 0.5)
y = (2 * x[:, 1] - 1 * x[:, 2] + 0.5 + 0.5 * np.random.randn(n)) # > 0
y = sign(y)
print(x.shape, y.shape)

lmd = 0.1 # 正則化項
delta = 0.01 # 勾配降下法の学習率
# x: n x 4 / y: n (vec) / w: 4

def J(w):
    global x, y
    return sum([math.log(1 + math.e**(-y[i] * np.dot(w.T, x[i]))) for i in range(n)]) + lmd * np.dot(w.T, w)

def dJdw(w):
    global x, y
    return sum([(-y[i] * x[i]) * math.e**(-y[i] * np.dot(w.T, x[i]))/(1 + math.e**(-y[i] * np.dot(w.T, x[i]))) for i in range(n)]) + 2 * lmd * w

def J_hessian(w):
    global x, y

def check(w):
    global x, y
    wrong = 0
    for i in range(n):
        y2 = sign(np.dot(w.T, x[i]))
        if y[i] != y2:
            wrong += 1
    return wrong
```

# 勾配降下法

```python
w = np.zeros(4)
```

```python
check(w)
```

```python
dJdw(w)
```
